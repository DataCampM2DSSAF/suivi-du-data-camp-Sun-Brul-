```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("/Users/romainbrule/Documents/Cours/Master2/KAGGLE/train_model.csv")
test=pd.read_csv("/Users/romainbrule/Documents/Cours/Master2/KAGGLE/test_final.csv")
```

```python
test.columns
id_seqpos=test.loc[:,['id_seqpos']]
```

## Modèle simple

```python
Y_train = train.loc[:,["reactivity","deg_Mg_pH10","deg_pH10","deg_Mg_50C","deg_50C"]]
X_train=train.loc[:,[ 'C', 'G', 'U','E', 'H', 'I', 'M', 'S', 'X', ')', 
             '.']]

X_test=test.loc[:,['C','G', 'U', 'E', 'H', 'I', 'M', 'S', 'X', ')', '.']]
```

```python
reg = LinearRegression()
reg.fit(X_train,Y_train)
y_pred = reg.predict(X_test)
```

```python
y_pred_df = pd.DataFrame(y_pred, columns= ['reactivity', 
            'deg_Mg_pH10', 'deg_pH10', 'deg_Mg_50C', 'deg_50C'])

y_pred_df.insert(0, "id_seqpos", id_seqpos, True) 

y_pred_df
```
